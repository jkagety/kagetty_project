# 1. Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import sqlite3
```

# 2. Loading the data

```python
conn=sqlite3.connect('im.db')
cursor = conn.cursor()
movie_budget_revenues = pd.read_csv('tn.movie_budgets.csv')
movie_basics=pd.read_sql("""SELECT* FROM movie_basics;""",conn)
movie_ratings =pd.read_sql("""SELECT* FROM movie_ratings;""",conn)
```

# 3. Joining the data together

```python
combined = pd.merge(movie_ratings, movie_basics, how='left')
combined  = combined.rename(columns={'original_title': 'movie'})#This changes the column name from original title to movie
final_table = pd.merge(movie_budget_revenues, combined, on='movie', how='left')
final_table
```

3.2 Dropping irrelevant columns

```python
columns_to_drop=['id','movie_id','numvotes','primary_title','start_year']
df=final_table.drop(columns_to_drop,axis=1)
df
```

3.3 Checking the data types of each column

3.3(i) converting the release date column from object to datetime datatype

```python
df.info()
df['release_date']=pd.to_datetime(df['release_date'])#This converts release date into a date time datatype
df
```

3.3(ii) Adding a new column month which is obtained from release date column

```python
df['Month'] = df['release_date'].dt.month
df
```

```python
months_in_year = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
df['Month'] = df['Month'].replace(months_in_year)#This labels the month column to their respective months
df
```

3.3(iii) Changing the data types of production budget, domestic gross and international gross from string to integer

```python
df['production_budget'] = df['production_budget'].str.replace('$', '').str.replace(',','')
df['production_budget'] = pd.to_numeric(df['production_budget'])
df['production_budget']
```

```python
df['worldwide_gross'] = df['worldwide_gross'].str.replace('$', '').str.replace(',','')
df['worldwide_gross'] = pd.to_numeric(df['worldwide_gross'])
df['worldwide_gross']
```

```python
df['domestic_gross'] = df['domestic_gross'].str.replace('$', '').str.replace(',','')
df['domestic_gross'] = pd.to_numeric(df['domestic_gross'])
df['domestic_gross']
```

# 4. Cleaning the data

```python
df.isna().sum() #Checks the number of missing values
columns_to_drop = ['averagerating', 'runtime_minutes', 'genres']
df.dropna(subset=columns_to_drop, inplace=True)#drops the rows containing the missing values
df
```

```python
df.isna().sum()#This checks for any missing values. The output is 0 which now proves there are no missing values in the dataframe
```

```python
df.duplicated().sum() #Checking for the duplicated values in the data. There are no duplicated values in the dataframe
```

Now that the data is clean and ready for analysis we go to our next step

6. Data Decisions
Were going to create a new column for profit.This will help answer which genres are more profitable.

```python
df['Profit'] = df['worldwide_gross'] - df['production_budget']
df
```

# 4 Exploratory Data Analysis And Visualisations

```python
#WHICH GENRE APPEARS MOST FROM THE DATA SET? WE ARE GOING TO RANK THE TOP TEN
grouped_data = df.groupby('genres').size().sort_values(ascending=False).reset_index(name='count')
top_10_genres_appearing =grouped_data.head(10)
top_10_genres_appearing =top_10_genres_appearing.reset_index(drop=True)
top_10_genres_appearing.index =top_10_genres_appearing.index + 1
top_10_genres_appearing
```

```python
# Bar graph for top 10 genres appearing
plt.figure(figsize=(10, 6))
bar_colors = ['red', 'yellow', 'blue','green','black','orange','violet','indigo','grey','pink']
plt.barh (top_10_genres_appearing['genres'], top_10_genres_appearing['count'],color=bar_colors)
plt.xlabel('Count')
plt.ylabel('Genres')
plt.title('most_frequented_genre')
```

```python
# Group movies by genres and the average ratings.
grouped_genre_averagerating = df.groupby('genres')['averagerating'].mean().reset_index()
sorted_genre_averagerating = grouped_genre_averagerating.sort_values(by='averagerating', ascending=False)
top_10_genres_averagerating= sorted_genre_averagerating.head(10)
top_10_genres_averagerating = top_10_genres_averagerating.reset_index(drop=True)
top_10_genres_averagerating.index = top_10_genres_averagerating.index + 1
top_10_genres_averagerating
```

```python
# Bar graph between for top 10 genres and average rating
plt.figure(figsize=(10, 6))
bar_colors = ['red', 'yellow', 'blue','green','black','orange','violet','indigo','grey','pink']
plt.barh(top_10_genres_averagerating['genres'], top_10_genres_averagerating['averagerating'],color=bar_colors)
plt.xlabel('averagerating')
plt.ylabel('Genres')
plt.title('Top 10 Genres that are highly rated')
plt.ticklabel_format(style='plain', axis='x')
```

```python
#Grouping genres by profit to answer which genre is the most profitable
grouped_genre_profit = df.groupby('genres')['Profit'].mean().reset_index()
sorted_genre_profit = grouped_genre_profit.sort_values(by='Profit', ascending=False)
top_10_sorted_genre= sorted_genre_profit.head(10)
top_10_sorted_genre = top_10_sorted_genre.reset_index(drop=True)
top_10_sorted_genre.index = top_10_sorted_genre.index + 1
top_10_sorted_genre
```
